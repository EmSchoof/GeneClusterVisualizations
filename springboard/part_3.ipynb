## Part 3 ‐ Predictive Modeling

**Background**: Ultimate is interested in predicting rider retention. To help explore this question, a sample dataset of a cohort of users who signed up for an Ultimate account in January 2014 has been provided (ultimate_data_challenge.json). The data was pulled several months later; consider a user retained if they were “active” (i.e. took a trip) in the preceding 30 days.

**Goal**: Use this data set to help understand what factors are the best predictors for retention, and offer suggestions to operationalize those insights to help Ultimate. *NOTE: Delete the dataset when you have finished with the challenge.*

Data description
- city: city this user signed up in
- phone: primary device for this user
- signup_date: date of account registration; in the form ‘YYYYMMDD’
- last_trip_date: the last time this user completed a trip; in the form ‘YYYYMMDD’
- avg_dist: the average distance in miles per trip taken in the first 30 days after signup
- avg_rating_by_driver: the rider’s average rating over all of their trips
- avg_rating_of_driver: the rider’s average rating of their drivers over all of their trips
- surge_pct: the percent of trips taken with surge multiplier > 1
- avg_surge: The average surge multiplier over all of this user’s trips
- trips_in_first_30_days: the number of trips this user took in the first 30 days after
signing up
- ultimate_black_user: TRUE if the user took an Ultimate Black in their first 30 days;
FALSE otherwise
- weekday_pct: the percent of the user’s trips occurring during a weekday

```python
# Load challenge json dataset
challenge_json_data = open('ultimate_data_challenge.json')
challenge_data = json.load(challenge_json_data)
# Define Data Type
type(challenge_data)
```

```python
# Assess Contents of Data
challenge_data[0]
```

*Observation*: Each list entry contains a list of value:key pairs
    - Key: shared labels
    - Value: unique values across list entries
*Solution*: 
    - each list item needs to be converted to row in the dataframe
    - each key is a row

```python
# Create a general function that:

#appends value to list
def add_to_list(li, val):
    return li.append(val)

#converts a list to series
def list_to_series(li):
    return pd.Series(li)
```

```python
# Grab key names for column
column_names = []

for each in challenge_data[0]:
    add_to_list(column_names, each)

# Test output
column_names
```

```python
ultimate_df = pd.DataFrame(challenge_data, columns=column_names)
ultimate_df.head()
```

```python
# Assess Dataframe
ultimate_df.info()
```

*Observations*: Most of the columns appear to be clean aside from 'avg_rating_of_driver', 'phone', and 'avg_rating_by_driver'. 

*Solution*: Drop all NaN in dataframe.

```python
# Drop NaN
ultimate_df = ultimate_df.dropna()
ultimate_df.info()
```

*Observations*: Columns 'signup_date' and 'last_trip_date' are numerical values in string format. 

*Solution*: Convert columns to datetime objects.

```python
# Import necessary modules
from datetime import datetime
```

```python
# Convert date strings to datetime objects

# signup
ultimate_df['signup_date'] = [datetime.strptime(x, '%Y-%m-%d') for x in ultimate_df['signup_date']]

# last_trip
ultimate_df['last_trip_date'] = [datetime.strptime(x, '%Y-%m-%d') for x in ultimate_df['last_trip_date']]

# Test output
ultimate_df[['signup_date','last_trip_date']].info()
```

```python
# Import necessary modules
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns; sns.set(style="darkgrid")
```

#### Categorical Values

```python
# Plot a histogram of categorical values
ultimate_df['ultimate_black_user'].value_counts().plot.bar(title="Frequency Distribution of Status of Ultimate Black User")
print('TRUE if the user took an Ultimate Black in their first 30 days; FALSE otherwise')
```

#### Numerical and Datetime Values

```python
# Grab key names for numeric column
numeric_cols = []
datetime_cols = []

for col in ultimate_df:
    
    # Define data type
    data_type = ultimate_df[col].dtype
    
    # If numeric
    if data_type == 'int64' or data_type == 'float64':
         add_to_list(numeric_cols, col)
    
    # If datetime
    elif data_type == 'datetime64[ns]':
        add_to_list(datetime_cols, col)

# Test output
print(numeric_cols, len(numeric_cols)) 
print('.')
print(datetime_cols)
```

```python
fig, ax = plt.subplots(len(numeric_cols), figsize=(10,12))

for i, col_val in enumerate(numeric_cols):

    sns.distplot(ultimate_df[col_val], bins=34, ax=ax[i], vertical=False, hist_kws={"rwidth":0.75,'edgecolor':'black', 'alpha':1.0})
    ax[i].set_title('Frequency Distribution '+ col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
    ax[i].set_ylabel('Count', fontsize=8)

plt.show()
```

*Observations*: 

**Skewed to the Left -> Indicates most values lie in the lower range**
- *trips_in_first_30_days* (number of trips this user took in the first 30 days after signing up)
- *avg_surge* (average surge multiplier over all of this user’s trips)
- *surge_pct* (percent of trips taken with surge multiplier > 1)
- *avg_dist* (average distance in miles per trip taken in the first 30 days after signup)

Theory: Users have low numbers of riders in the first month of using rideshares, and price surges, especially above 1, appear to be uncommon.

**Skewed to the Right -> Indicates most values lie in the higher range**
- *avg_rating_by_driver* (rider’s average rating over all of their trips)
- *avg_rating_of_driver* (rider’s average rating of their drivers over all of their trips)

Theory: A passenger's and driver's ratings seem to be largely high values.

**Wide Distribution Across range -> peaks in high and low values**
- *weekday_pct* (percent of the user’s trips occurring during a weekday) 

Theory: Some users only use rideshares on non-weekdays (low value peak), some users use rideshares on both weekdays and non-weekdays (low, wide range in middle), and some users only use rideshare on weekdays (high value peak).
